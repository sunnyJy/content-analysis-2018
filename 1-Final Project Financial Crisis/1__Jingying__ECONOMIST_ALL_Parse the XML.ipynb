```python
import xml.etree.ElementTree as ET
import pandas #gives us DataFrames
import re #for regexs
import numpy as np #math calculation
import glob
```

```python
years = ['2006','2007','2008','2009','2010']
months = ['01','02','03','04','05','06','07','08','09','10','11','12']
YEAR = []
MONTH = []
DAY = []

for y in years:
    for m in months:
        day_list = glob.glob('ECON-'+y+'-'+m+'*.xml')
        for dl in day_list:
            d = re.sub('ECON-'+y+'-'+m, '',dl)
            d = re.sub('_Text.xml','',d)
            DAY.append(d)
            MONTH.append(m)
            YEAR.append(y)
    print('finish year '+y+' month '+m)

Calendar = pandas.DataFrame({'Year': YEAR,
                        'Month': MONTH,
                        'Day': DAY})
Calendar = Calendar[['Year','Month','Day']]
```

```python
def perXml(y,m,d):
    global ECON_YEAR
    global ECON_MONTH
    global ECON_DAY
    global ECON_TITLE
    global ECON_TAG
    global ECON_SECTOR
    global ECON_TEXT
    global ECON_Article_Index
    
    YEAR = []
    MONTH = []
    DAY = []
    TITLE = []
    TAG = []
    SECTOR = []
    TEXT = []
    Article_Index = []
    
    
    #tree = ET.parse('/Users/techbar/Downloads/ECON-2007-0106_Text.xml')
    tree = ET.parse('/Users/techbar/Downloads/ECON-'+y+'-'+m+d+'_Text.xml')
    root = tree.getroot()
    
    Topics = ['Leaders', 'Britain', 'Europe', 'United States', 'The Americas', 
              'Middle East and Africa', 'Asia', 'Obituary',
              'Business', 'Finance and economics', 'Science and technology', 'Books and arts']
    
    #Topics = ['Business', 'Finance and economics', 'Science and technology', 'Books and arts']
    
    
    # content_num
    M = root.findall('artInfo')
    search_content = 0
    count = 0
    for mm in M:
        n = mm.find('ocrText').text
        if search_content == 0:
            count += 1
            print(str(count))
            if re.search('Contents', n):
                ocrText = n
                search_content = 1
    
    
    #artInfo = root.findall('artInfo')[content_num]
    #ocrText = artInfo.find('ocrText').text
    P = ocrText.split('\n\t\t\t')

    for topic in Topics:
        for p in P:
            if re.findall(topic,p) != []:
                title_list = re.split(r' \d+ ',p)
                tag = title_list[0]
                for tl in range(1,len(title_list)):
                    TITLE.append(title_list[tl])
                    for t in Topics:
                        if re.search(t, tag):
                            TAG.append(t)
                        else:
                            TAG.append(tag)
            
                    YEAR.append(y)
                    MONTH.append(m)
                    DAY.append(d)
                    
                    if topic == 'Science and technology':
                        SECTOR.append('Science & Tech')
                    elif topic == 'Books and arts':
                        SECTOR.append('Books & Arts')
                    else:
                        SECTOR.append('Finance & Econ')


    for i in range(0, len(TITLE)):
        find = 0
        temp_text = []
        temp_index = []
        
        for j in range(count+1,len(root.findall('artInfo'))):
            A = root.findall('artInfo')[j]
            B = A.find('ocrText').text


            T = re.sub(r'[?|$|)|!|(|@|#|%|^|&|*]',r'',TITLE[i])
            #nstr = re.sub(r'[?|$|.|!]',r'',strs)

            word_bag = T.split(' ')
            total_num = len(word_bag)
            count = 0
            for w in range(0,len(word_bag)):
                try: 
                    if re.search(word_bag[w], B):
                        count += 1
                except:
                    print('This is article in year' + y + ' month ' + m + ' day ' + d)
                    print('The article title is ' + TITLE[i])
                    print('tag is ' + tag)
            if count >= total_num-2:
                
            #if re.search(T,B):
                #print(TITLE[i])
                #print(B)
                #print('****************************')
                temp_text.append(B)
                temp_index.append(j)
                find = 1
        if find == 0:
            temp_text.append('n.a.')
            temp_index.append('n.a.')
        TEXT.append(temp_text[len(temp_text)-1])
        Article_Index.append(temp_index[len(temp_index)-1])
        
        
    ECON_YEAR.extend(YEAR)
    ECON_MONTH.extend(MONTH)
    ECON_DAY.extend(DAY)
    ECON_TITLE.extend(TITLE)
    ECON_TAG.extend(TAG)
    ECON_SECTOR.extend(SECTOR)
    ECON_TEXT.extend(TEXT)
    ECON_Article_Index.extend(Article_Index)
    
    return
```

```python
ECON_YEAR = []
ECON_MONTH = []
ECON_DAY = []
ECON_TITLE = []
ECON_TAG = []
ECON_SECTOR = []
ECON_TEXT = []
ECON_Article_Index = []

for i in range(0,len(YEAR)):
    y = YEAR[i]
    m = MONTH[i]
    d = DAY[i]
    perXml(y,m,d)
    
    if int(YEAR[i])-int(YEAR[i-1]) != 0:
        print('start year '+YEAR[i])
    if int(MONTH[i])-int(MONTH[i-1]) != 0:
        print('start month '+MONTH[i])    
```
